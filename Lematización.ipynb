```python
import re
import unicodedata as ud
from nltk.tokenize import word_tokenize 

def remove_punctuation(words):
    """Remove punctuation from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub(r'[^\w\s\À-ú]', '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words

texto = ["""está protección investigación y tecnología """]
texto2 = ['holá qúé']

print(remove_punctuation(texto2))

def replace_all(dict, str):
    for key in dict:
        str = str.replace(key, dict[key])
    return str

good_text = []
for t in texto:
    u = replace_all({"á":"á", "é":"é", "í":"í", "ó":"ó", "ú":"ú"}, t)
    good_text.append(u)
    
#print(good_text)
print(remove_punctuation(good_text))

articles = []

for article in good_text:
    new_article = word_tokenize(article)
    article_words = []
    for e in new_article:
        if ud.category(e) == 'Ll':
            article_words.append(e)
    articles.append(article_words)
    
```
